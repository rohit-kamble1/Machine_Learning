```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
os.chdir("E:\pandas")
```

```python
df=pd.read_csv("Breast_Cancer.csv")
df.head()
```

```python
dummies=pd.get_dummies(df.diagnosis)
dummies.head()
```

```python
merged= pd.concat([df, dummies], axis='columns')
merged.tail()
```

```python
final_df= merged.drop(['diagnosis','Unnamed: 32','M',], axis='columns')
final_df.head()
```

```python
final_df.corr().transpose()
```

```python
X=final_df[["radius_mean","texture_mean"]]
y=final_df[["B"]]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test,y_train,y_test= train_test_split(X,y, test_size=0.3)
```

```python
from sklearn import tree
clf= tree.DecisionTreeClassifier()
```

```python
clf.fit(X_train, y_train)
```

```python
y_pred= clf.predict(X_train)
```

```python
from sklearn.metrics import classification_report

classification_report(y_train, y_pred)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_train, y_pred)
```

```python
#MLP Classifier
from sklearn.neural_network import MLPClassifier
clf=MLPClassifier(hidden_layer_sizes=(100,100,100), alpha=0.0001)
```

```python
clf.fit(X_train, y_train)
```

```python
y_pred=clf.predict(X_train)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_train, y_pred)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_train, y_pred)
```

```python
from sklearn.metrics import classification_report
classification_report(y_train, y_pred)
```
